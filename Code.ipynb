```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

from sklearn.datasets import load_breast_cancer

from sklearn import metrics

import numpy as np
import pandas as pd
```

```python
cancer = load_breast_cancer()
```

```python
algorithms = [KNeighborsClassifier(),SVC()]
```

Average values of metrics after 5-fold cross-validation

```python
for algorithm in algorithms:
    print(algorithm)
    print(np.average(cross_val_score(algorithm, cancer.data, cancer.target,cv=5)),"Accuracy")
    print(np.average(cross_val_score(algorithm, cancer.data, cancer.target,cv=5, scoring='recall')),"Recall")
```
